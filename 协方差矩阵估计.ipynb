## 5. Shrinkage estimators
The essential idea is that the unbiased but often poorly estimated sample covariance can be combined with a structured estimator ùêπ
, using the below formula (where ùõø is the shrinkage constant):
 
                                    Œ£ÃÇ=ùõøùêπ+(1‚àíùõø)ùëÜ
It is called shrinkage because it can be thought of as ‚Äúshrinking‚Äù the sample covariance matrix towards the other estimator, which is accordingly called the shrinkage target. The shrinkage target may be significantly biased but has little esimation error. There are many possible options for the target, and each one will result in a different optimal shrinkage constant ùõø. PyPortfolioOpt offers the following shrinkage methods:
* Ledoit-Wolf shrinkage:
    * constant_variance shrinkage, i.e the target is the diagonal matrix with the mean of asset variances on the diagonals and zeroes elsewhere. 
    * single_factor, Based on Sharpe‚Äôs single-index model which effectively uses a stock‚Äôs beta to the market as a risk model. 
    * constant_correlation shrinkage, in which all pairwise correlations are set to the average correlation (sample variances are unchanged). 

```python
model=risk_models.CovarianceShrinkage(price_data)
cov_data=model.ledoit_wolf(shrinkage_target='constant_correlation')
cov_data
```

* Oracle approximating shrinkage (OAS), invented by Chen et al. (2010) [7], which has a lower mean-squared error than Ledoit-Wolf shrinkage when samples are Gaussian or near-Gaussian.

```python
model=risk_models.CovarianceShrinkage(price_data)
cov_data=model.oracle_approximating()
cov_data
```
