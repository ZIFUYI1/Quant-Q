## 3. exponentially weighted covariance

α[p_0+(1−α)p_1+(1−α)^2*p_2+…+(1−α)^n*p_n+…]

```python
cov_data=risk_models.exp_cov(price_data, span=180, frequency=252)
cov_data
```

## 5. Shrinkage estimators
The essential idea is that the unbiased but often poorly estimated sample covariance can be combined with a structured estimator 𝐹
, using the below formula (where 𝛿 is the shrinkage constant):
 
                                    Σ̂=𝛿𝐹+(1−𝛿)𝑆
It is called shrinkage because it can be thought of as “shrinking” the sample covariance matrix towards the other estimator, which is accordingly called the shrinkage target. The shrinkage target may be significantly biased but has little esimation error. There are many possible options for the target, and each one will result in a different optimal shrinkage constant 𝛿. PyPortfolioOpt offers the following shrinkage methods:
* Ledoit-Wolf shrinkage:
    * constant_variance shrinkage, i.e the target is the diagonal matrix with the mean of asset variances on the diagonals and zeroes elsewhere. 
    * single_factor, Based on Sharpe’s single-index model which effectively uses a stock’s beta to the market as a risk model. 
    * constant_correlation shrinkage, in which all pairwise correlations are set to the average correlation (sample variances are unchanged). 

```python
model=risk_models.CovarianceShrinkage(price_data)
cov_data=model.ledoit_wolf(shrinkage_target='constant_correlation')
cov_data
```

* Oracle approximating shrinkage (OAS), invented by Chen et al. (2010) [7], which has a lower mean-squared error than Ledoit-Wolf shrinkage when samples are Gaussian or near-Gaussian.

```python
model=risk_models.CovarianceShrinkage(price_data)
cov_data=model.oracle_approximating()
cov_data
```
